```python
import pandas as pd
import numpy as np
from collections import Counter

def loadLabor(address):
    """读取数据并对‘？’进行替换"""
    spf = pd.read_csv(address, sep=',',index_col = False, header = 0)
    # 已经将上一行代码中的参数header的值设为了0，则下面不需要对列名进行赋值了
#     column = ['duration','wage-increase-first-year','wage-increase-second-year','wage-increase-third-year',
#              'cost-of-living-adjustment','working-hours','pension','standby-pay','shift-differential','education-allowance',
#              'statutory-holidays','vacation','longterm-disability-assistance','contribution-to-dental-plan',
#               'bereavement-assistance','contribution-to-health-plan','class']
#     spf.columns = column
    str_typeName = ['cost-of-living-adjustment','pension','education-allowance',
                   'vacation','longterm-disability-assistance','contribution-to-dental-plan',
                   'bereavement-assistance','contribution-to-health-plan','class']

    # 建立替换字典
    str2numeric = {}
    str2numeric['?'] = '-1'
    spf = spf.replace(str2numeric) # 替换
    return spf, str2numeric,str_typeName

def fillMissData(spf, str_typeName):
    row, col = spf.shape
    columns = spf.columns
    for column_name in columns:
        # 若不在标签型数据列表里，则使用数值型，即使用均值填充
        if column_name not in str2numeric:
            tmp = spf[column_name].apply(float)
            ave = np.average(tmp[tmp!=-1])
            tmp[tmp==-1] = ave
            spf[column_name] = tmp
        else:  # 若在标签型数据列表内，则使用众数填充
            v = spf[column_name].values
            v1 = v[v != '-1']
            c = Counter(v1)
            cc = c.most_commotn(1)
            v[v == '-1'] = cc[0][0]
        return spf

if __name__ == '__main__':
    filepath = 'labor.txt'
    fillFilePath = 'laborMissing_handle.txt'
    spf,str2numeric,str2numeric = loadLabor(filepath)
    spf = fillMissData(spf,str2numeric)
    spf.to_csv(fillFilePath, index = False, header = False)
```
